```python
import pandas as ps
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from numpy.random import randn
```

```python
sns.get_dataset_names()
```

```python
ds=sns.load_dataset('tips')
ds.head()
```

```python
ds.describe()
```

# relplot

```python
sns.relplot(x='total_bill',y='tip' ,data=ds)
```

# line plot

```python
x=ds['total_bill'].head()
y=ds['tip'].head()
plt.plot(x,y,'g--')
```

# histograme

```python
plt.hist(x)
```

# barplot

```python
plt.barh(x,y)
```

# Pie chart

```python
plt.pie(x,labels=y)
```

```python
sns.relplot(x='total_bill',y='tip' ,data=ds,style='time',hue='size')
```

```python
sns.relplot(x='total_bill',y='tip' ,data=ds,hue='time',style='size',size='smoker')
```

```python
fmri=sns.load_dataset('fmri')
```

```python
fmri
```

# scatterplot

```python
sns.scatterplot(x='subject',y='signal', data=fmri,hue='region' ,style='event')
```

# catplot

```python
sns.catplot(x='region',y='signal', data=fmri,kind='swarm')
```

```python
sns.catplot(x='signal',y='region', data=fmri,hue='subject',kind='box',dodge='false')
```

```python
sns.catplot(x='signal',y='region', data=fmri.sort_values('subject'),hue='event',kind='boxen',dodge='false')
```

```python
sns.catplot(x='signal',y='region', data=fmri.sort_values('subject'),hue='event',kind='violin',dodge='false')
```
